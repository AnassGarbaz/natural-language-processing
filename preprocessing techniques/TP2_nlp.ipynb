# Q1 : 

---
For this question we need to import stopwords and use them to count the sw in our data

---

```
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
stopWords = set(stopwords.words('english'))
```

--- 

we need to tokenize our texts for the processing.

---

```
from nltk.tokenize import  word_tokenize
nltk.download('punkt')
```

```
#this list will contain all counts of stop words in the 6 documents.
all_stopWordsCount = []
#this list will contain all counts of tokens in the 6 documents.
all_ToknCount = []
#this list will contain all counts of upper case tokens in the 6 documents.
all_ToknCountMaj = []
for i, doc in enumerate(webtext.fileids()) : 
  #tokenize the doc
  words = word_tokenize(webtext.raw(doc))
  cStopWords = 0
  CountTokn = 0
  CountToknMaj = 0
  for w in words:
    # counting tokens
    CountTokn +=1
    if w in stopWords:
      #counting stop words
      cStopWords += 1
    if w.isupper():
      #counting upper case tokens
      CountToknMaj += 1
  #we add the values of every doc in this lists for results
  all_stopWordsCount.append(cStopWords)
  all_ToknCount.append(CountTokn)
  all_ToknCountMaj.append(CountToknMaj)
```

```
# We put the results in a dict
data = {'Doc name': webtext.fileids(),
         'Stop Words' : all_stopWordsCount,
        'Tokens' : all_ToknCount,
        'Maj Tokens' : all_ToknCountMaj}
```

```
import pandas as pd
```

---

Making the Dataframe of data dict.

---

```
df = pd.DataFrame(data)
print(df)
```

# Q2 : 

---

Removing stop words and upper case tokens.

---

```
# this list will contain the new 6 docs with no stop words or upper case tokens
Docs_NoSW_NoMaj = []

for i, doc in enumerate(webtext.fileids()) : 
  # tokenize the doc
  words = word_tokenize(webtext.raw(doc))
  for i, w in enumerate(words):
    if w in stopWords:
      #if w is a stop word delete it
      words.remove(w)
    if w.isupper():
      #if w is uc make it lower case
      words[i]= w.lower()
  #putting the new doc in the list
  Docs_NoSW_NoMaj.append(words)
```

Deleting punctuation and numbers from the docs.

---

```
import re
import string 
```

```
# Storing the sets of punctuation in variable result  
result = string.punctuation  
    
# Printing the punctuation values  
print(result)  
```

---

this function delete punc and numb from a doc given.

---

```
def Del_Punct_Numb(doc):
    for i, stri in enumerate(doc):
      # if the tok is a punct delete it 
       if stri in result : 
         doc.remove(stri)
       #if there is a numb or numb with punct delete it 
       else : 
          st = re.sub(r'[^\w\s]|[0-9]','',stri)
          doc[i] = st
    # deleting the blank element in the doc list.
    doc = [w for w in doc if w != '']
    return doc
```

Testing

```
s = [".0.", 'an2od;', ".2",'0.',"25.", '.0','054..',"1487/*.",'babyBo2']

print(Del_Punct_Numb(s))
```

```
import emoji
import regex
```

---
This function take a text as a parameter and return number of emojis.

---

```
def CountEmojis(text):
    #emoji counter
    count = 0
    #spliting the text 
    data = regex.findall(r'\X', text)
    for word in data:
        #if the word is an emoji add 1 to the counter.
        if any(char in emoji.UNICODE_EMOJI for char in word):
            count += 1

    return count
```

Testing.

```
text = "🤔 🙈 lorum ipsum 😌 de 💕👭👭"
count = CountEmojis(text)
print("Number of emojis is : {}".format(count))
```

#Q8 : 

---

 porter stemming application : 

---

```
from nltk.stem import PorterStemmer
```

```
# obtain stems for all docs
ps = PorterStemmer()
#list for all docs stems
pstems = []
for  doc in webtext.fileids() : 
   words = word_tokenize(webtext.raw(doc))
   #stem for one doc
   stem=[ps.stem(word) for word in words]
   pstems.append(stem)
```

```
print("stems for Doc 1 (firefox.txt) : ", pstems[0])
```

---

lemmatization with WordNetLemmatizer

---

```
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

lemmatization for all docs

```
wl = WordNetLemmatizer()

lemms = []
for  doc in webtext.fileids() : 
   words = word_tokenize(webtext.raw(doc))
   lemm = []
   lemm =[wl.lemmatize(word) for word in words]
   lemms.append(lemm)
```

lemmatization of doc 1 ( firefox.txt)

```
print(lemms[0])
```

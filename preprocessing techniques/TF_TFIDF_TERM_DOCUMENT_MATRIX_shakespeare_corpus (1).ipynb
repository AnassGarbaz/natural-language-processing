#Preparing the documents

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np 
from textblob import TextBlob
import nltk
from nltk.corpus import shakespeare #XMLCorpusreader
nltk.download('shakespeare')
nltk.download('punkt')
```

Set Documents : 

```
document_0 = "You are trying to code TF-IDF all by yourself like a big girl / boy."
document_1 = "So this is a tinny doc."
document_2 = "And another tinny doc to test few stuff."
document_3 = "So in total, we are four documents, have fun ;)."
all_documents = [document_0, document_1, document_2, document_3]
```

Make tokens for all words :  

```
all_documents_tokn = []
for doc in all_documents:
 #all_documents_tokn.append(nltk.tokenize.word_tokenize(doc))
 doc = TextBlob(doc)
 all_documents_tokn.append(doc.words)
```

```
print(all_documents_tokn[0])
```

#Q1 : TF

Count word in a document  

```
def countWordInDoc(w, d):
  c=0
  for word in d : 
     if word == w:
       c += 1
  return c
```

TF function : 

```
def TF(w, d): 
  return countWordInDoc(w, d) / len(d)
```

# Q2 : IDF

Counts word in the collection

```
def countCollectionsHadWord(w, D):
  con=0
  for doc in D : 
     if countWordInDoc(w, doc) != 0 :
       con+=1
  return con
```

IDF function

```
def IDF(w, D): 
  return np.log(len(all_documents_tokn) / countCollectionsHadWord(w, D))
```

# Q3  : TF-IDF

```
def TF_IDF(w, d, D):
  return TF(w, d) * IDF(w, D)
```

#Q5 : term document 

Select all words that we have in collection : 

```
all_words = []
for doc in all_documents_tokn : 
   for w in doc :
     all_words.append(w)
```

Select distinct strings from all words by changing it to a set 

```
setWords = set(all_words)
```

Reconvert the set to a list again : 

```
def convertSetToList(set):
  return [* set, ]
```

```
all_words = convertSetToList(set(all_words))
```

Make term document matrix :

```
data = {'name': all_words }
```

```

for i, doc in enumerate(all_documents_tokn) : 
  all_words_tfidf = []
  for  w in all_words:
        all_words_tfidf.append(TF_IDF(w, doc, all_documents_tokn))
  data.__setitem__('doc%d'% i , all_words_tfidf)
```

```
df_marks = pd.DataFrame(data)
print(df_marks)
```

Make bigrams for all lists :

```
sp_bigrams = []
for cor in shakespeare_corpus:
  sp_bigrams.append([(x, i.split()[j + 1]) for i in cor
       for j, x in enumerate(i.split()) if j < len(i.split()) - 1])
```

```
print(sp_bigrams)
```

Put all bigrams in one list to simplify the TF-IDF calculation : 

```
all_bigrams = []
for cor in sp_bigrams : 
   for w in cor :
     all_bigrams.append(w)
```

Select distinct bigrams :

```
setWords = set(all_bigrams)
def convertSetToList(set):
  return [* set, ]

all_bigrams = convertSetToList(set(all_bigrams))
```

The term document matrix for bigrams : 

```
data2 = {'bigrams': all_bigrams }
```

```
for i, doc in enumerate(sp_bigrams) : 
  all_words_tfidf = []
  for  w in all_bigrams:
        all_words_tfidf.append(TF_IDF(w, doc, sp_bigrams))
  data2.__setitem__('cor%d'% i , all_words_tfidf)
```

```
df2 = pd.DataFrame(data2)
print(df2)
```

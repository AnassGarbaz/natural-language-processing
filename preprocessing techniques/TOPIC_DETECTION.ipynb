```
from nltk.corpus import stopwords
import re
import string 
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

--------

Calling the stop words 

```

stopWords = set(stopwords.words('english'))
```

-----

Removing sw and punc and upper case tokens 

-----

```
# this list will contain the new 6 docs with no stop words or upper case tokens
Docs_NoSW_NoMaj = []
# Storing the sets of punctuation in variable result  
result = string.punctuation  

for i, doc in enumerate(brown.fileids()) : 
  # tokenize the doc
  words = word_tokenize(brown.raw(doc))
  for i, w in enumerate(words):
    if w in stopWords:
      #if w is a stop word delete it
      words.remove(w)
    if w.isupper():
      #if w is uc make it lower case
      words[i]= w.lower()
    if w in result : 
      #removing puntuation
         words.remove(w)
  #putting the new doc in the list
  Docs_NoSW_NoMaj.append(words)
```

# Q3 : 

----

Representing the 500 documents in matrix form

----

Making a corpus for all docs :

```
corpus = []
for doc in brown.fileids():
  corpus.append(brown.raw(doc))
```

-----
Displaying the results 

-----

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
dataVec = X.toarray()
print("Featues name \n ", vectorizer.get_feature_names())
print('the matrix \n ', dataVec)
```

#Q4 : 

----
Importing the models : 

----

```
from sklearn.decomposition import NMF, LatentDirichletAllocation, TruncatedSVD
```

-----

LatentDirichletAllocation

```
NUM_TOPICS = 10
# Build a Latent Dirichlet Allocation Model
lda_model = LatentDirichletAllocation(n_components=NUM_TOPICS, max_iter=10, learning_method='online')
lda_Z = lda_model.fit_transform(dataVec)
```

----

NMF


```
nmf_model = NMF(n_components=NUM_TOPICS)
nmf_Z = nmf_model.fit_transform(dataVec)
```

----

TruncatedSVD

```
lsi_model = TruncatedSVD(n_components=NUM_TOPICS)
lsi_Z = lsi_model.fit_transform(dataVec)
```

# Q5 : 

-----

display the first 10 words characterizing each topic

-----

```
def print_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i]) for i in topic.argsort()[:-top_n - 1:-1]])
 
print("LDA Model:")
print_topics(lda_model, vectorizer)
print("=" * 50)
 
print("NMF Model:")
print_topics(nmf_model, vectorizer)
print("=" * 50)
 
print("LSI Model:")
print_topics(lsi_model, vectorizer)
print("=" * 50)
```

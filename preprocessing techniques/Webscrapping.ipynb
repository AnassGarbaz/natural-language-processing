```python
import newspaper
from newspaper import Config
```

```python
newspaper.languages()
```

```python
config = Config()
hespress=newspaper.build(url='https://www.hespress.com/',language='ar', memoize_articles=False, config=config)
hespress.print_summary()
```

```python
#showing the articles urls
for article in hespress.articles:
    print(article.url)
```

```python
#first article
hes_article=hespress.articles[0]
print(hes_article.title)
#print(hes_article.fulltext)
print(hes_article.authors)
```
